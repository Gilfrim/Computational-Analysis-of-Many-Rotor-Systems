```python
import numpy as np
import matplotlib.pyplot as plt

from quant_rotor.core.dense.de_solve_one_thermal import integration_scheme
from quant_rotor.core.dense.de_solve_one_thermal_dense import integration_scheme as integration_scheme_fast
```

```python
site = 3
state = 11
g = 0.1
```

```python
beta_s, energy_s, t_1_max_s, t_2_max_s  = integration_scheme_fast(site, state, g, t_init=0,t_final=10,nof_points=100000)
```

```python
beta_d, energy_d, t_1_max_d, t_2_max_d  = integration_scheme(site, state, g, t_init=0,t_final=10,nof_points=100000)
```

```python
plt.plot(beta_s, -energy_s, label=f"TD Singles g = {g}")
plt.plot(beta_d, -energy_d, label=f"TD Singles g = {g}")
plt.title("Energy Imaginary Time Propagation (Os Basis)")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend()
plt.show()
```

```python
np.allclose(energy_d, energy_s, atol=1e-20)
```
